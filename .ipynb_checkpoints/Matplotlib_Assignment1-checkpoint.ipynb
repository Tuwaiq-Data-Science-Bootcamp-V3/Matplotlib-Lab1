## Q1: Create two charts using list and dataframe.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
edu = pd.read_csv('tmdb-movies.csv')
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
Newx = ['Riyadh','Jeddh','Makkah','Abha','Dammam']
Newy = [ 120,35,60,50,130]

plt.pie(Newy,labels=Newx, autopct='%0.1f%%', startangle=120, shadow=True, radius=1, explode = [0.1,0.2,0,0.2,0.1])

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
df = pd.DataFrame([x,y1,y2])
plt.scatter(df[0],df[1]) 
plt.xlabel(0,fontsize = 20)
plt.ylabel(1,fontsize = 20)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color

```python




plt.title('Example title',loc= 'left',color='red')
plt.scatter(df[0],df[1]) 
plt.xlabel("Y",fontsize = 60)
plt.ylabel("X",fontsize = 100)
plt.show()
plt.show()
```
