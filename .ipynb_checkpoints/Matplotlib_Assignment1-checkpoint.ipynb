## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(range(10,50,10), [500,550,430,300], c= 'blue')
plt.plot(range(1,10,1), range(10,19,1), c ='green')
plt.plot(range(21,31,1), range(31,41,1), c= 'black')
plt.legend(['First', 'Second', 'Third'], shadow= True, loc='upper left')
plt.xlabel('Rndom Numbers')
plt.ylabel('Random Number2')
plt.title('Line Charts of Random Number')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots', fontsize= 20 )


# First plot
plt.subplot(1,3,1)
plt.plot(range(10,50,10), [500,550,430,300], c= 'blue')
plt.title('Line Charts of Random Number')
plt.xlabel('Rndom Numbers')
plt.ylabel('Random Number2')
# Second plot
plt.subplot(1,3,2)
plt.plot(range(1,10,1), range(10,19,1), c ='green')
plt.title('Line Charts of Random Number')
plt.xlabel('Rndom Numbers')
plt.ylabel('Random Number2')
#Third plot
plt.subplot(1,3,3)
plt.plot(range(21,31,1), range(31,41,1), c= 'black')
plt.title('Line Charts of Random Number')
plt.xlabel('Rndom Numbers')
plt.ylabel('Random Number2')
plt.show()
```
