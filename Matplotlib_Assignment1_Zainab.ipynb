## Q1: Create two charts using list and dataframe.

```python
import numpy as np
import pandas as pd

dataFrame = pd.DataFrame({'X': [2008,2009,2010,2011,2012,2013],
                    'Y': [180, 200, 270, 330, 312, 550]})

listDataX = [2008,2009,2010,2011,2012,2013]
listDataY = [130,210,240,480,540,630]
  

dataFrame
```

```python
plt.figure(figsize=(5,3))

plt.plot(dataFrame["X"],dataFrame["Y"], c ='cornflowerblue', linestyle = 'dashed', linewidth = 2)
plt.plot(listDataX,listDataY, c ='orange', linestyle = 'solid', linewidth = 2)
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Number of Orders every Year')
plt.legend(['Jeddah Branch','Riyadh Branch'], shadow = True, loc = 0)
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.figure(figsize=(5,3))

branches = ['Jeddah', 'Riyadh']
ordersNum = [dataFrame["Y"].sum(), sum(listDataY)]

plt.title('Number of Orders every Year')
plt.pie(ordersNum,labels=branches, autopct='%1.1f%%', startangle=90, shadow=True, radius=1, explode = [0.06,0.06], colors=["cornflowerblue","orange"])

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.figure(figsize=(5,3))

plt.scatter(dataFrame["X"], dataFrame["Y"], alpha = 0.4, marker = 'o', c="cornflowerblue", s = 50, label = 'Jeddah Orders') # s >> size
plt.scatter(listDataX, listDataY, alpha = 0.4, marker = 'o', c="orange", s = 50, label ='Riyadh Orders')
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(figsize=(5,3))

plt.scatter(dataFrame["X"], dataFrame["Y"], alpha = 0.4, marker = 'o', c="cornflowerblue", s = 50, label = 'Jeddah Orders') # s >> size
plt.scatter(listDataX, listDataY, alpha = 0.4, marker = 'o', c="orange", s = 50, label ='Riyadh Orders')

plt.legend(['Jeddah Branch','Riyadh Branch'], shadow = True, loc = 0)
plt.title('Number of Orders every Year', fontsize = 16, c="red", loc="left")
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Number of Orders', fontsize = 16)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(dataFrame["X"],dataFrame["Y"], c ='cornflowerblue', linestyle = 'dashed', linewidth = 2)
plt.plot(listDataX,listDataY, c ='orange', linestyle = 'solid', linewidth = 2)
plt.plot([2008,2009,2010,2011,2012,2013], [93, 160, 70, 210, 270, 206], c ='cyan', linestyle = 'dotted', linewidth = 2)
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Number of Orders every Year')
plt.legend(['Jeddah Branch','Riyadh Branch', 'Hail Branch'], shadow = True, loc = 0)
plt.show()
```
