```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
lst = [1, 2, 3]
data = {
  "A": [4, 6, 9]
}

df = pd.DataFrame(data)
print (lst)
print (df)
```

```python
plt.plot(range(1,5), [1,2,3,4])
plt.plot(lst,df, c ='green', linestyle = 'solid', linewidth = 3)
plt.xlabel('range')
plt.ylabel('Number from 1 to 10')
plt.title('Q1')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
genres = ['Comedy','Drama','Comedy|Romance','Adventure|Action|Thriller','Comedy|Drama|Romance']
sizes = [ 20323889661,15107533347,11545724341,7597483556,7251175696]

plt.pie(sizes,labels=genres, autopct='%1.1f%%', startangle=90, shadow=True, radius=1, explode = [0.3,0.3,0.3,0.3,0.3])

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df["A"],lst, alpha = 0.4, s=50) 

plt.title('Q3',fontsize = 10)
plt.xlabel('alpha =0.4',fontsize = 10)
plt.ylabel('size= 50',fontsize = 10)


plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.plot(range(1,5), [1,2,3,4])
plt.plot(lst,df, c ='green', linestyle = 'solid', linewidth = 3)
plt.legend(['df','lst'], shadow = True, loc = 0)
plt.xlabel('range',fontsize = 16)
plt.ylabel('Number from 1 to 10',fontsize = 16)
plt.title('Q1' ,color='r', loc='left')
plt.show()
```
