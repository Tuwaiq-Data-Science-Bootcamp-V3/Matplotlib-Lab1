```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
# LIST
year = ["2010", "2012", "2014", "2016"]
rav = [100,111, 199, 160]

# DataFrame
data = {'Year': ["2010", "2012", "2014", "2016"],
        'Sales': [60, 50, 160, 200]}

# Creating a dataframe
df = pd.DataFrame(data)
```

```python

# Creating the line chart

plt.plot(year, rav)
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Line Chart')
plt.show()
 
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(rav,labels=year, autopct='%1.1f%%', startangle=180, shadow=True, radius=1, explode = [0.3,0.3,0.3,0.3])

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Year'],df['Sales'], alpha = 0.4, marker = 'o', s = 50) 

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['Year'],df['Sales'], alpha = 0.4, marker = 'o', s = 50) 
plt.title('Year vs. Sales',fontsize = 16, c ="r", loc='left')
plt.xlabel('Year',fontsize = 16)
plt.ylabel('Sales',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
year2=["2010","2012","2014","2015"]
sales2 =[80,30,78,8]


plt.plot(year, rav , color ='#A52A2f')
plt.plot(df['Year'], df['Sales'],  color ='cyan')
plt.plot(year2,sales2, color ='gree')

plt.legend(['classA','classB','classC'], shadow = True, loc = "upper left")
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('3 charts',fontsize = 40)



plt.plot(year, rav)
plt.plot(df['Year'], df['Sales'])
plt.plot(year2,sales2)


#first
plt.subplot(3,1,1) 
plt.pie(rav,labels=year, autopct='%1.1f%%')
plt.title('Pie chart' ,loc="left");

#second
plt.subplot(3,1,2)
plt.plot(year, rav)
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Line Chart' ,loc="left")


#third
plt.subplot(3,1,3)
plt.scatter(df['Year'],df['Sales'])
plt.title('scatter chart', loc="left")
plt.xlabel("Year")
plt.ylabel('Sales'); 


```
