```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
mylist=[8,9,7,10,9] #created list

data = {'Name': ['Amal', 'Sara', 'Hanadi', 'Noura', 'Lama'],'marks': [7, 9, 6, 8, 10]}
df = pd.DataFrame(data, index=['1','2','3','4','5']) #created dataframe

Name= df['Name']
Marks= df['marks']

plt.plot(Name, mylist)
plt.plot(Name,Marks, c ='k', linewidth = 3)
plt.xlabel('Name of student')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(Marks, labels=Name, autopct='%1.f%%', colors=["r", "g", "b", "c","y"], explode = [0.3,0.2,0,0,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(Name,Marks, alpha = 0.4, s=50) 

plt.title('Marks of Students')
plt.xlabel('Students Names')
plt.ylabel('Marks')

plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(Name,Marks, alpha = 0.4, s=50) 

plt.title('Marks of Students', loc='left', c='r')
plt.xlabel('Students Names',fontsize =16)
plt.ylabel('Marks', fontsize = 16)

plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(Name,[2,4,6,8,10],'b');
plt.plot(Name,mylist,'r');
plt.plot(Name,Marks,'g');
plt.legend(['1','2','3'], shadow = True, loc = 2);
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])

plt.subplot(3,1,1)
plt.scatter(Name,Marks, alpha = 0.4, s=50) 
plt.title('Marks of Students')
plt.xlabel('Students Names')
plt.ylabel('Marks');

plt.subplot(3,1,2)
plt.plot(Name,[2,4,6,8,10],'b')
plt.plot(Name,mylist,'r')
plt.plot(Name,Marks,'g')
plt.legend(['1','2','3'], shadow = True, loc = 2);

plt.subplot(3,1,3)
plt.plot(Name, mylist)
plt.plot(Name,Marks, c ='k', linewidth = 3)
plt.xlabel('Name of student')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show()
```
