## Q5: Create three line charts with Legends at left and colors.

```python
plt.figure(figsize=(10,6))
plt.plot(np.arange(1, 20), c="r", label="Line1")
plt.plot(np.arange(4, 16), c="b", label="Line2")
plt.plot(np.arange(2, 30, 3), c="g", label="Line3")
plt.legend(loc="upper left")
plt.show()
```
